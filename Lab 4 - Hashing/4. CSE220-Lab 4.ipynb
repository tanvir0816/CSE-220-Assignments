```python
#Complete this constructor
class vehicleNodes:
  def __init__(self, brand, vehicle_type, rent, passenger, next = None):
    #TO DO
    self.brand=brand
    self.vehicle_type=vehicle_type
    self.rent=rent
    self.passenger=passenger
    self.next=next
```

```python
class VehicleHashTable:
    def __init__(self, size):
        self.vehicleTable = [None]*size
        self.empty_slot = {}
        for i in range(size):
          self.empty_slot[i] = False

    def create_from_vehicle_info_array(self, arr):
        for i in arr:
            self.insert_vehicle(i)

    def print_vehicle_hashtable(self):
        idx = 0
        for i in self.vehicleTable:
            print(idx, ':', end = ' ')
            head = i
            while head != None:
                print(f'(Brand: {head.brand}, Type: {head.vehicle_type}, Rent: {head.rent}, No. of Passengers: {head.passenger})', end = '---->')
                head = head.next
            print('None')
            print()
            idx += 1

    def find_empty_slot(self):
        for k,v in self.empty_slot.items():
            idx = k
            del self.empty_slot[k]
            return idx


  #Do it by yourself
    def __hash_function(self, brand):
        hehe=0
        for i in brand:
            hehe+=ord(i)
        index=hehe%len(self.vehicleTable)

      # Do not discard the next 3 lines
        if index in self.empty_slot:
            del self.empty_slot[index]
        return index


  #Do it by yourself
    def insert_vehicle(self, vehicle):
        #TO DO

        br=vehicle[0]
        vtype=vehicle[1]
        rent=vehicle[2]
        pas=vehicle[3]
        idx=self.__hash_function(br)
        x=vehicleNodes(br,vtype,rent,pas)
        if self.vehicleTable[idx] is None:
            self.vehicleTable[idx]=x
        else:

            if self.vehicleTable[idx].brand!=br:
                self.vehicleTable[self.find_empty_slot()]=x
            else:

                x.next=self.vehicleTable[idx]
                self.vehicleTable[idx]=x

        return self.vehicleTable
```

Implement the __hash_function() and insert() methods

```python
class Node:
  def __init__(self, value=None, next = None):
    self.value = value
    self.next = next

class HashTable:
  def __init__(self, length):
    n = Node()
    self.ht = [n] * length
    self.length = length

  def show(self):
    count = 0
    for i in self.ht:
      temp = i
      print(count, end=" ")
      while temp!=None:
        print (temp.value, end="-->")
        temp = temp.next
      count += 1
      print()


  #Do it by yourself
  def __hash_function(self, key):
    #TO DO
    xx=len(key)
    sum=0
    if xx%2==0:
        for i in range(xx):
            if i%2==0:
                sum+=ord(key[i])
    else:
        for i in range(xx):
            if i%2!=0:
                sum+=ord(key[i])
    idx=sum%len(self.ht)
    return idx


  #Do it by yourself
  def insert(self, key, value):
    #TO DO
    node=Node((key,value))
    indx=self.__hash_function(key)
    curr=self.ht[indx]
    if curr.value == None:
        self.ht[indx]=node
    else:
        if  node.value[1]>self.ht[indx].value[1]:
            node.next=curr
            self.ht[indx]=node
        else:
            prev = None
            while curr != None and node.value[1] < curr.value[1]:
                prev = curr
                curr = curr.next
            if prev:
                prev.next = node
                node.next = curr
            else:
                node.next = self.ht[indx]
                self.ht[indx] = node



    return self.ht
```

```python
class Node_pair:
  def __init__(self, key, value, next = None):
    self.key, self.value, self.next = key, value, next


class Hashtable:
  def __init__(self, size):
    self.ht = [None]*size


  def insert(self, s):
    hashed_index = self.__hash_function(s[0])
    pair = Node_pair(s[0], s[1])
    if self.ht[hashed_index] == None:
      self.ht[hashed_index] = pair
    else:
      pair.next = self.ht[hashed_index]
      self.ht[hashed_index] = pair


  def create_from_array(self, arr):
    for i in arr:
      self.insert(i)

  def print_hashtable(self):
    idx = 0
    for i in self.ht:
      print(idx, ':', end = ' ')
      head = i
      while head != None:
        print(f'({head.key}, {head.value})', end = '-->')
        head = head.next
      print('None')
      idx += 1


  #Do it by yourself
  def __hash_function(self, key):
    #TO DO
    key=int(key)
    idx=(key+3)%len(self.ht)
    return idx
  #Do it by yourself
  def remove(self, key):
    #TO DO
        hashed_index = self.__hash_function(str(key))
        current = self.ht[hashed_index]
        prev = None

        while current:
            if current.key == key:
                if prev:
                    prev.next = current.next
                else:
                    self.ht[hashed_index] = current.next
                return
            prev = current
            current = current.next

        print(f"Key {key} not found.")


```
